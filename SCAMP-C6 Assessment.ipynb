# DATA PREPROCESSING


```python
#checking for missing values
df.isnull().sum()
```

```python
df.isnull().sum()
#checking missing values again
```

```python
#dataset imformation after dropping the missing values
df.info()
```

```python
#the dataset contains numbers and text or categorical data so the data is splitted into two
df_num=df[['job_id','telecommuting','has_company_logo','has_questions','fraudulent']]
#df_num contains the numerical values in the dataset
```

```python
#df_cat contains the categorical(text) values in the dataset
df_cat=df[['title','location','department','salary_range','company_profile','description','requirements','benefits','employment_type','required_experience','required_education','industry','function']]
```

```python
df_num
```

```python
#using visuals to check for outliers in the numeric value
plt.figure(figsize=[16,8])
sns.boxplot(data=df_num)
plt.show()
#the diagram below shows the presence of outliers in "telecommuting","has_company_logo","fraudulent"
```

```python
#dealing with the outliers
df_num=df_num[df_num['telecommuting']<0.9]
df_num=df_num[df_num['has_company_logo']<0.1]
df_num=df_num[df_num['fraudulent']<0.9]
```

```python
df_num
```

```python
#checking again to confirm that the outliers have been dealt with
plt.figure(figsize=[16,8])
sns.boxplot(data=df_num)
plt.show()
```

# INDUSTRY WITH THE HIGHEST NUMBER OF FAKE JOBS


```python
#a new variable was created to only contain the part of the dataset that is fraudulent that is fake
df_industry=df[df['fraudulent']==1]
```

```python
df_industry.shape
#the row and column of the fraudulent dataset
```

```python
df_industry['industry'].value_counts()
#this the number of unique values in the column industry of the fraudulent variable
```

```python
#INDUSTRY WITH THE HIGHEST NUMBER OF FAKE JOBS
plt.subplots(figsize=(15,10))
sns.countplot('industry',data=df_industry)
plt.xticks(rotation=90)
plt.title('INDUSTRY WITH THE HIGHEST NUMBER OF FAKE JOBS')
plt.show()
```

```python
print('Industry with the highest number of fake job: ',df_industry['industry'].value_counts().index[0])
```
